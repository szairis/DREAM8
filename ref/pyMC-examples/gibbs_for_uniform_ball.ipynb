Uniform points in an $n$-dimensional ball=========================================This notebook implements a Gibbs sampler in PyMCto sample uniformly from an $n$-dimensional ball,i.e to sample from the set$$\mathbf{B}_n = \\{x \in \mathbf{R}^n: \|x\|\leq 1\\}$$

```python
# simple modeln = 5X = [mc.Uninformative('X_%d'%i, value=0) for i in range(n)]@mc.potentialdef in_ball(X=X):    if pl.dot(X, X) <= 1.:        return 0    else:        return -pl.inf
```

```python
class UniformBall(mc.Gibbs):    def __init__(self, stochastic, others, verbose=None):        self.others = others        self.conjugate = True # pymc will include a Metropolis rejection step on top of the proposal if this is false        mc.Gibbs.__init__(self, stochastic, verbose)        def propose(self):        x_other = [X_i.value for X_i in self.others]        max_val = pl.sqrt(1. - pl.dot(x_other, x_other))        self.stochastic.value = mc.runiform(-max_val, max_val)
```

```python
m = mc.MCMC([X, in_ball])for i in range(n):    m.use_step_method(UniformBall, X[i], [X[j] for j in range(n) if j != i])
```

```python
time m.sample(1000, progress_bar=False)
```

```python
def plot_trace(k=500):    pl.figure(figsize=(2*n,2*n))    for i in range(n):        for j in range(i):            pl.subplot(n, n, i*n+j+2)            pl.plot(X[i].trace()[:k], X[j].trace()[:k], 'k.-', alpha=.5)            pl.xticks([])            pl.yticks([])        try:            pl.title('$X_%d$'%j, fontsize=32)        except UnboundLocalError:            pass        pl.subplot(n, n, i*n+1)        pl.plot(X[i].trace()[:k], 'k-', alpha=.9)        pl.xticks([])        pl.yticks([])        pl.ylabel('$X_%d$'%i, fontsize=32, rotation='horizontal')    pl.subplots_adjust(hspace=0, wspace=0)    plot_trace()
```

```python
def plot_acorr(k=1000):    pl.figure(figsize=(2*n,2))    for i in range(n):        pl.subplot(1, n, i+1)        pl.acorr(X[i].trace()[:k])        pl.xticks([])        pl.yticks([])        pl.title('$X_%d$'%i, fontsize=32)        pl.axis([-11,11,-.1, 1.1])    pl.subplots_adjust(hspace=0, wspace=0)    plot_acorr()
```

"Proof" of rapid mixing of Gibbs step=====================================I know that the Metropolis step is "rapidly mixing" in the sense that a number of steps bounded by a polynomialfunction of the number of dimensions will be within $\epsilon$ of the stationary distribution.But I know that the Gibbs-step chain gets near its stationary distribution much more rapidly.But knowing these things is very far from proving these things.  I've always wanted to have a simple proof.

My favorite path to a simple proof of rapid mixing is the coupling method.In this setting, it would take the form of coupling a chain $(X^1, X^2, \ldots)$ with a chain $(Y^1, Y^2, \ldots)$,so that the marginal distribution of the transition probability density for each chain is the same as the transitionprobability density of the original chain.Then if the expected distance between $X^N$ and $Y^N$ is small, there should be a rigorous statement about how closethe chain is to its stationary distribution after $N$ steps.Here is how to do it for the Gibbs-step chain:* Pick a coordinate axis $i_t$ uniformly at random* Pick a sign $s_t \in \\{-1,1\\}$ uniformly at random* Pick a proportion $p_t \in [0,1]$ uniformly at random* Update $X$ and $Y$ by moving coordinate $i_t$ to $s_t \cdot p_t \cdot d_t$, where $d_t$ is the maximum distancethat this coordinate can be from the axis, given all of the other coordinate valuesClear?  In other words, $X^t_j = X^{t-1}_j$ and $Y^t_j = Y^{t-1}_j$ for all $j \neq i_t$, and$$X^t_{i_t} = s_t p_t \sqrt{1 - \sum_{j\neq i_t} X^t_j^2}$$and$$Y^t_{i_t} = s_t p_t \sqrt{1 - \sum_{j\neq i_t} Y^t_j^2}.$$Claim 1 is that the marginal distribution of $X$ and the marginal distribution of $Y$ are both the same as the Gibbs-step chain above.Claim 2, which takes more work, is that the expected distance between $X^t$ and $Y^t$ tends rapidly to zero.

```python
class CoupledBall(mc.Gibbs):    def __init__(self, X, Y, verbose=None):        assert len(X) == len(Y), 'need same dimension to couple'        self.n = len(X)        self.i = 0        self.X = X        self.Y = Y        self.conjugate = True # pymc will include a Metropolis rejection step on top of the proposal if this is false        mc.StepMethod.__init__(self, X+Y, verbose)        def propose(self):        self.i = mc.np.random.randint(self.n) # random scan        #self.i = (self.i + 1) % self.n # sequential scan        s = [-1,1][mc.np.random.randint(2)]        p = mc.np.random.rand()                #print i, s, p        for Z in [self.X, self.Y]:            Z_other = [Z_j.value for j, Z_j in enumerate(Z) if j != self.i]            d_Z = pl.sqrt(1. - pl.dot(Z_other, Z_other))            Z[self.i].value = s*p*d_Z                        #print d_Z, Z[i].value
```

```python
n = 25X = [mc.Uninformative('X_%d'%i, value=0.) for i in range(n)]Y = [mc.Uninformative('Y_%d'%i, value=0.) for i in range(n)]dist = mc.Lambda('dist', lambda X=X, Y=Y: pl.rms_flat(pl.array(X)-pl.array(Y)))
```

```python
# get initial values for X and Y that are differentfor Z in [X, Y]:    m = mc.MCMC(Z)    for i in range(n):        m.use_step_method(UniformBall, Z[i], [Z[j] for j in range(n) if j != i])    %time m.sample(1000, progress_bar=False)
```

```python
m = mc.MCMC([X, Y, dist])m.use_step_method(CoupledBall, X, Y)
```

```python
time m.sample(1000, progress_bar=False)
```
